```python
import numpy as np

existing_array = np.array([[1, 2, 3], [4, 5, 6]])
array_with_border = np.pad(existing_array, pad_width=1, mode='constant', constant_values=0)

print("Existing Array:")
print(existing_array)
print("\nArray with Border:")
print(array_with_border)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Assuming diabetes_df is your DataFrame with Pima Indians Diabetes data

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
diabetes_df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')

# g) Correlation and scatter plots
correlation_matrix = diabetes_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

# Scatter plots for selected pairs
sns.scatterplot(x='Glucose', y='Insulin', data=diabetes_df)
plt.title('Scatter Plot: Glucose vs Insulin')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()

# h) Histograms
diabetes_df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of Pima Indians Diabetes Data', y=0.95)
plt.show()

# i) Three-dimensional plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(diabetes_df['Glucose'], diabetes_df['Insulin'], diabetes_df['BMI'], c='blue', marker='o')
ax.set_xlabel('Glucose')
ax.set_ylabel('Insulin')
ax.set_zlabel('BMI')
ax.set_title('Three-dimensional Scatter Plot')
plt.show()
```

Task 2: 9. Compare the results of the Univariate and Bivariate analysis for the UCI diabetes data set

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming diabetes_df is your DataFrame with UCI diabetes data

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
# diabetes_df = pd.read_csv('your_dataset.csv')

# Univariate Analysis

# Histograms for each variable
diabetes_df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Univariate Analysis - Histograms', y=0.95)
plt.show()

# Bivariate Analysis

# Pairplot for selected variables
sns.pairplot(diabetes_df[['Pregnancies', 'Glucose', 'Insulin', 'BMI', 'Outcome']], hue='Outcome', markers=['o', 's'], diag_kind='kde')
plt.suptitle('Bivariate Analysis - Pairplot', y=1.02)
plt.show()

# Correlation Matrix
correlation_matrix = diabetes_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Bivariate Analysis - Correlation Matrix')
plt.show()
```


Univariate Analysis:
Histograms:Reveals the distribution of individual variables like Pregnancies, Glucose, Insulin, BMI, etc.Helps identify patterns, skewness, and central tendencies in each variable.
Bivariate Analysis:
Pairplot:
Displays scatter plots for pairs of variables, colored by the outcome (diabetic or not).Helps observe relationships between variables and potential separations between diabetic and non-diabetic groups.Diagonal KDE plots show the distribution of each variable individually.
Correlation Matrix:
Quantifies the linear relationship between pairs of variables.The heatmap visually represents the strength and direction of correlations.Useful for identifying potential collinearity and understanding how variables interact.
Comparison:
Univariate analysis is essential for understanding the characteristics of each variable independently.Bivariate analysis extends the understanding by exploring relationships between variables, especially regarding the outcome (diabetic or not).Both analyses provide complementary insights, aiding in identifying key features and potential predictors of diabetes.
Observations:
The histograms reveal the distribution of each variable, such as a positively skewed distribution for Glucose.In the pairplot, we observe patterns and separations between diabetic and non-diabetic groups.The correlation matrix indicates relationships between variables, where, for example, Glucose is positively correlated with Outcome.
