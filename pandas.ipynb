ISNULL() AND NOTNULL() FUNCTION IN PANDAS

DATAFRAMING THE DATA BY MELT()

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [90, 85, 92],
    'Physics': [88, 89, 95]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Chemistry': [75, 80, 88],
    'Biology': [92, 87, 78]
})

df1_melted = pd.melt(df1, id_vars=['ID', 'Name'], var_name='Subject', value_name='Score')
print("Melted DataFrame:")
print(df1_melted)
```

DATAFRAMING THE DATA BY MERGE()

```python
merged_df = pd.merge(df1_melted, df2, on='ID')
print("\nMerged DataFrame:")
print(merged_df)
```

MEAN(),MEDIAN(),AND STANDARD DEVIATION() OF AN DATA IN PANDAS

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Calculating the mean, median, and standard deviation for each column
mean_values = df.mean()
median_values = df.median()
std_dev_values = df.std()

# Displaying the calculated statistics
print("\nMean Values:")
print(mean_values)

print("\nMedian Values:")
print(median_values)

print("\nStandard Deviation Values:")
print(std_dev_values)
```

CHANGING DATA IN PANDAS BY APPLY() AND MAP()

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'Salary': [50000, 60000, 45000, 70000],
    'Department': ['HR', 'IT', 'Finance', 'IT']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Using apply() to apply a function to a specific column
def double_salary(salary):
    return salary * 2

df['Doubled_Salary'] = df['Salary'].apply(double_salary)
print("\nDataFrame with Doubled Salary:")
print(df)
```

```python
# Using map() to map values in a column to new values
department_mapping = {'HR': 'Human Resources', 'IT': 'Information Technology', 'Finance': 'Finance'}
df['Department_Full'] = df['Department'].map(department_mapping)

print("\nDataFrame with Mapped Department Names:")
print(df)
```

DATA CLEANING BY RENAME() AND DROP() IN PANDAS

```python
data = {
    'Original_Name': [1, 2, 3, 4],
    'Value': [11, 22, 33, 44],
    'Another_Column': [111, 200, 333, 400]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# Using rename() to rename a column
df = df.rename(columns={'Original_Name': 'New_Name'})
print("\nDataFrame with Renamed Column:")
print(df)
```

```python
# Using drop() to remove a column
df = df.drop(columns=['Another_Column'])
print("\nDataFrame with Column Removed:")
print(df)
```


 PLOTTING THE DATAFRAME USING MATPLOTLIB

```python
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-01-10'),
    'Value_A': [25, 30, 22, 35, 28, 33, 40, 45, 38, 42],
    'Value_B': [15, 18, 20, 25, 22, 30, 35, 40, 32, 38]
}

df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

# Plotting the DataFrame using Matplotlib
df.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Line Plot of Values over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```

```python
# Creating a sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value_1': np.random.randint(1, 100, 5),
    'Value_2': np.random.randint(1, 100, 5)
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Bar plot
df.plot(kind='bar', x='Category', y=['Value_1', 'Value_2'], figsize=(10, 6))
plt.title('Bar Plot of Values by Category')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend(['Value_1', 'Value_2'])
plt.grid(axis='y')
plt.show()

# Scatter plot
df.plot(kind='scatter', x='Value_1', y='Value_2', s=100, c='red', figsize=(8, 6))
plt.title('Scatter Plot of Value_1 vs. Value_2')
plt.xlabel('Value_1')
plt.ylabel('Value_2')
plt.grid(True)
plt.show()
```
